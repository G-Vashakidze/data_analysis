```python
import pandas as pd
file_path = 'data.csv'
df = pd.read_csv(file_path)
```

```python
df
mean_ladder_score = df["Ladder score"].mean()
std_dev_ladder_score = df["Ladder score"].std()
median_ladder_score = df["Ladder score"].median()
min_ladder_score = df["Ladder score"].min()
max_ladder_score = df["Ladder score"].max()
print(mean_ladder_score, std_dev_ladder_score, median_ladder_score, min_ladder_score ,max_ladder_score)
```

```python
columns_to_exclude = ["Standard error of ladder score", "upperwhisker","lowerwhisker","Dystopia + residual","Ladder score in Dystopia"]
selected_columns_df = df.drop(columns=columns_to_exclude)
```

```python
selected_columns_df
```

```python
d = selected_columns_df  
d.set_index('Country name')
```

```python

import seaborn as sns
import matplotlib.pyplot as plt

sns.lmplot( x='Healthy life expectancy',y='Ladder score', data=d, height=10)
plt.title("Relationship between Happiness Index and Healthy Life Expectancy")
plt.ylabel("Happiness Index")
plt.xlabel("Healthy Life Expectancy")

plt.show()
```

```python

ladder_score_mean = d["Ladder score"].mean()
gdp_mean = d["Logged GDP per capita"].mean()

ladder_score_filter = d["Ladder score"] >= ladder_score_mean
gdp_filter = d["Logged GDP per capita"] <= gdp_mean

filtered_df = d[ladder_score_filter & gdp_filter]

filtered_df

```

```python
ladder_score_mean = d["Ladder score"].mean()
gdp_mean = d["Logged GDP per capita"].mean()

ladder_score_filter = d["Ladder score"] <= ladder_score_mean
gdp_filter = d["Logged GDP per capita"] >= gdp_mean

filtered_df = d[ladder_score_filter & gdp_filter]

filtered_df
```

```python


sns.lmplot(x='Logged GDP per capita', y='Healthy life expectancy', data=d, height=10)
plt.title("gdp and Healthy Life Expectancy")
plt.xlabel("Logged GDP per capita")
plt.ylabel("Healthy Life Expectancy")

plt.show()
```

```python

sns.lmplot(x="Ladder score", y='Logged GDP per capita', data=d, height=12)
plt.title("gdp and Healthy Life Expectancy")
plt.xlabel("Logged GDP per capita")
plt.ylabel("Healthy Life Expectancy")

plt.show()
```

```python
import numpy as np

countries = filtered_df["Country name"]
ladder_score = filtered_df["Ladder score"]

x_indices = np.arange(len(countries))

bar_width = 0.6
plt.bar(x_indices, ladder_score, width=bar_width)
plt.xlabel("Countries")
plt.ylabel("Ladder Score")
plt.title("Ladder Score by Country (Filtered Data)")
plt.xticks(x_indices, countries, rotation=90)
plt.show()


```

```python

countries = filtered_df["Country name"]
gdp_per_capita = filtered_df["Logged GDP per capita"]


x_indices = np.arange(len(countries))


plt.plot(x_indices, gdp_per_capita, marker='o', linestyle='-')
plt.xlabel("Countries")
plt.ylabel("Logged GDP per capita")
plt.title("Logged GDP per Capita by Country (Filtered Data)")
plt.xticks(x_indices, countries, rotation=90)
plt.show()
```
